```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier
from io import StringIO
import pydotplus
from ipywidgets import Image
```

```python
tree = DecisionTreeClassifier(random_state=17)
tree.fit(X, y)
```

```python
dot_data = StringIO()
export_graphviz(tree, feature_names=X.columns, out_file=dot_data, filled=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(value=graph.create_png())
```

```python
from sklearn import preprocessing 
label = preprocessing.LabelEncoder() 
data = pd.read_csv('income_evaluation.csv', sep=',')
for y in data.columns:
    if(data[y].dtype == object):
        data[y]= label.fit_transform(data[y])
X = data
X = X.drop([' income'], axis='columns')
y = data[' income']
X.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y)
```

```python
dtree = RandomForestClassifier().fit(X_train, y_train)
```

```python
feats = {}
for feature, importance in zip(X.columns, dtree.feature_importances_):
    feats[feature] = importance
print(feats)
```

```python
dtree.score(X_test, y_test)
```

```python
b_dtree=BaggingClassifier(RandomForestClassifier(), n_estimators=5, random_state=42)
b_dtree.fit(X_train, y_train)
b_dtree.score(X_test, y_test)
```

```python
import numpy as np
from catboost import CatBoostClassifier
from sklearn.metrics import mean_squared_error
scores = np.ndarray([10, 10])
for i, n_trees in enumerate(np.linspace(1, 100, 10)):
    for j, depth in enumerate(np.arange(2, 12)):
        p = CatBoostClassifier(n_estimators=n_trees, max_depth=depth, task_type="GPU").fit(X_train, y_train).predict(X_test)
        scores[i, j] = mean_squared_error(y_test, p)
```

```python
import matplotlib.pyplot as plt
for i in range(scores.shape[0]):
    plt.plot(np.linspace(1, 100, 10), scores[i], label='depth' + str(np.arange(2, 12)[i]))
plt.legend()
```

```python
for i in range(scores.shape[1]):
    plt.plot(np.arange(2, 12), scores[i], label='trees=' + str(np.linspace(1, 100, 10)[i]))
plt.legend()
```
